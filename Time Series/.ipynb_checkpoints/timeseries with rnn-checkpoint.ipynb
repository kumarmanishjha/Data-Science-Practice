```python
#Read data in the form of dates as index. In most of the other datasets, we convert dates into features, 
#however here we are using only the patter in the data itself to predict the number of passesngers in future.
dateparse = lambda dates: dt.strptime(dates, '%Y-%m')
data = pd.read_csv('AirPassengers.csv', parse_dates=['Month'], index_col='Month',date_parser=dateparse)
data.head()
```

```python
#Determing moving statistics
#Sometimes also known as running or rolling statistics
#Time period is 12 to signify 12 months of one year
moving_avg = data.rolling(12).mean()
moving_std = data.rolling(12).std()

#Plot the series and statistics
plt.plot(data, color='blue',label='Original Series')
plt.plot(moving_avg, color='red', label='Moving average')
plt.plot(moving_std, color='purple', label = 'Moving Std. deviation')
plt.legend(loc='best')
plt.title('Moving average & Standard Deviation')
plt.show()
```

```python
#Scale data w.r.t. moving average and moving standard deviation
data = data.iloc[:,0]
moving_avg = moving_avg.iloc[:,0]
moving_std = moving_std.iloc[:,0]
scaled_data = (data - moving_avg)/moving_std
```

```python
#Plot the series and statistics
plt.plot(scaled_data, color='blue',label='Original Series')
plt.title('Scaled data using moving statistics')
plt.show()
```

# Approach to use RNN for timeseries predictions

1. Create dataset by randomly selecting sequences of random sizes from the original training data

```python
#Generate data for training RNN
train_len = scaled_data.size
N = 1000 #Number of sequences to sample
seq_len = 12
max_start_pos = train_len-seq_len-1  #Maximum Start position from original data to sample a sequence
use_noise = False
```

```python
all_sequences = []
target = []
for i in range(N):
    #Sample start index
    start_index = np.random.randint(0,max_start_pos)
    
    #Calculate end index
    end_index = min(train_len-2, start_index + seq_len)
    
    #Get sequence by subsetting from original data
    seq = np.array(scaled_data.iloc[start_index:end_index])
    target.append(scaled_data.iloc[end_index+1])
    #print(start_index)
    #print(end_index)
    #print(seq)
    
    #Store in all_sequences
    all_sequences.append(seq)
```

```python
target = torch.tensor(target, dtype=torch.float)
target.size()
```

```python
class Model(nn.Module):
    def __init__(self, input_size=1, hidden_layer_size=100, output_size=1):
        
        super(Model, self).__init__()
        
        self.hidden_layer_size = hidden_layer_size

        self.lstm = nn.LSTM(input_size, hidden_layer_size)
        
        self.linear = nn.Linear(hidden_layer_size, output_size)

        self.linear = nn.Linear(hidden_layer_size, output_size)
        
        self.hidden_cell = (torch.zeros(1,1,self.hidden_layer_size),
                            torch.zeros(1,1,self.hidden_layer_size))

    def forward(self, input_seq):
        lstm_out, self.hidden_cell = self.lstm(input_seq.view(len(input_seq) ,1, -1), self.hidden_cell)
        predictions = self.linear(lstm_out.view(len(input_seq), -1))
        return predictions[-1]
```

```python
model = Model()
loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
print(model)
```

```python
model = model.to(device)
```

```python
def train_model(dataset,target, n_epoch = 100):
    for i in range(n_epoch):
        for seq, label in zip(dataset, target):
            #print(seq)
            #print(label)
            seq = seq.to(device)
            label = label.to(device)
            
            optimizer.zero_grad()
            
            model.hidden_cell = ((torch.zeros(1, 1, model.hidden_layer_size)).to(device), 
                                 (torch.zeros(1, 1, model.hidden_layer_size)).to(device))
            y_pred = model(seq)

            single_loss = loss_function(y_pred, label)
            single_loss.backward()
            optimizer.step()

        if i%25 == 1:
            print(f'epoch: {i:3} loss: {single_loss.item():10.8f}')

    print(f'epoch: {i:3} loss: {single_loss.item():10.10f}')
```

```python
train_model(train_data,target, n_epoch = 100)
```
